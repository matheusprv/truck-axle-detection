```python
import os 
```

```python
train_folder = r"Dataset\train\labels"
files_path = [ os.path.join(train_folder, x) for x in os.listdir(train_folder)]
```

```python
data = []
number_axles_per_file = []
for file in files_path:
    with open(file, 'r') as f:
        lines = f.readlines()
        if len(lines) == 0:
            print(file)
        number_axles_per_file.append(len(lines))
        for line in lines:
            x, y, w, h = line.split(" ")[1:]
            data.append((float(x), float(y)))
```

```python
import matplotlib.pyplot as plt

# Unpack the list of tuples into x and y coordinates
x, y = zip(*data)

# Create the plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Centro dos eixos')  # Scatter plot


# Add labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Localizao do centro dos eixos')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
mean = sum(number_axles_per_file) / len(number_axles_per_file)
maximum = max(number_axles_per_file)
minimum = min(number_axles_per_file)
print(mean, maximum, minimum)

num_axles = {k: 0 for k in range(minimum, maximum+1)}
for num in number_axles_per_file:
    num_axles[num] += 1

print(num_axles)
```

```python
# Dados
indexes = list(num_axles.keys())
values = list(num_axles.values())

# Criando o grfico de barras
plt.figure(figsize=(5, 3))
plt.bar(indexes, values, width=0.6, edgecolor='black', alpha=0.7)
plt.title("Frequncia da quantidade de eixos")
plt.xlabel("Quantidade de eixos")
plt.ylabel("Frequncia")
plt.xticks(indexes)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Exibindo o grfico
plt.show()
```
